```python
data = {'name': ['Joe', 'Bill', 'Alice'],
'salary': [5000, 6000, 8000]}
```

```python
df_salary
```

```python
# give 10% raise to every employee
temp = df_salary.loc[:,'salary'] * 1.10
```
